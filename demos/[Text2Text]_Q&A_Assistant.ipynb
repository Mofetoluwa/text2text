```python
import requests

def get_data(lang_code="en"):
  url = f"https://raw.githubusercontent.com/deepmind/xquad/master/xquad.{lang_code}.json"
  r = requests.get(url)
  d = r.json()
  return [p["context"].lower() for a in d["data"] for p in a["paragraphs"]]
```

```python
import text2text as t2t
#lang_codes = ["en","es","de","el","ru","tr","ar","vi","th","zh","hi","ro"]
corpus = get_data(lang_code="en")
index = t2t.Handler(corpus).index(ids=list(range(len(corpus))))
```

```python
query = "who did tesla partner with?"
dist, pred_ids = index.search([query.lower()], k=1)
corpus[pred_ids[0][0]]
```

```python
## Without knowledge base, Assistant uses pretrained knowledge, which may be out of context
t2t.Handler([query]).assist()
```

```python
## Without knowledge base, Assistant responds specifically using the context
t2t.Handler([query]).assist(knowledge_base=(corpus,index))
```
