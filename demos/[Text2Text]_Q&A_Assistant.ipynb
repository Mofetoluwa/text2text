```python
import requests

def get_data(lang_code="en"):
  url = f"https://raw.githubusercontent.com/deepmind/xquad/master/xquad.{lang_code}.json"
  r = requests.get(url)
  d = r.json()
  return [p["context"].lower() for a in d["data"] for p in a["paragraphs"]]
```

```python
import text2text as t2t
#lang_codes = ["en","es","de","el","ru","tr","ar","vi","th","zh","hi","ro"]
corpus = get_data(lang_code="en")
index = t2t.Indexer().transform(corpus)
query = "who did tesla partner with?"
matching_docs = index.retrieve([query.lower()], k=1)
print(matching_docs)
```

```python
## Assistant uses pretrained knowledge if no knowledge base provided
t2t.Assistant().transform([query])
```

```python
## Assistant uses intended context if knowledge base retriever provided
t2t.Assistant().transform([query], retriever=index)
```
